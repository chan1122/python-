```python
# 삼성전자 데이터 가져오고
STOCK_CODE = '005930'
```

```python
# 변수 stock에 삼성전자 데이터를 할당하고 값을 확인합니다
stock = fdr.DataReader(STOCK_CODE)
```

```python
# 1990년~2000년, 2000년~2010년, 2010년~2015년, 2015년~2020년으로 나뉘어진 4개의 그래프를 만들기 시작합니다.
time_steps = [['1990','2000'],
              ['2000','2010'],
              ['2010','2015'],
              ['2015','2020']]

# 행과 열이 2개가 되도록 설정을 해주면~~~!
fig, axes = plt.subplots(2,2)
fig.set_size_inches(16, 9)
for i in range(4):
  ax = axes[i//2, i%2]
  df = stock.loc[(stock.index > time_steps[i][0])&(stock.index < time_steps[i][1])]
  sns.lineplot(y=df['Close'], x=df.index, ax=ax)
  ax.set_title(f'{time_steps[i][0]}~{time_steps[i][1]}')
  ax.set_xlabel('time')
  ax.set_ylabel('price')
plt.tight_layout()
plt.show()
```

```python
# 예측을 위해 사용할 값을 stock이라는 변수에 할당해줍니다
stock = fdr.DataReader(STOCK_CODE, '2022')
```

```python
# 그리고 우리는 프로펫을 사용하여 값을 예측하는 기능을 추가할 것인데요~!
from fbprophet import Prophet
from fbprophet.plot import plot_plotly, plot_components_plotly
```

```python
# y라는 컬럼을 만들어서 Close에 있는 값을 똑같이 넣어주세요
# ds라는 컬럼을 만들어서 Date에 데이터를 같은 과정으로 넣어준다면~!
stock['y'] = stock['Close']
stock['ds'] = stock.index
```

```python
# 컬럼을 확인햇을떄 y 컬럼과 ds 컬럼이 생긴 것을 확인할 수 있죠~!
stock.head()
```

```python
# 변수 m에 프로펫을 할당해주즈는 것이애요~!
m = Prophet()
# 위에서 선언한 stock를 학습시켜야되니 아래 코드를 쳐줍니다~!
m.fit(stock)
```

```python
# 예측을 어느 정도까지 할 것인지 ~~~
# 현재 코드에서는 30일까지만 예측을 할 것이기때문에 (periods=이분에 30을 넣습니다~!)
future = m.make_future_dataframe(periods=30)

# future.tail()을 이용하여 어느정도까지 예측을 할 것인지 다시 한 번 날짜를 확인할 수 있습니다.
# future.tail() 을 이용한다면 최종 예측일부터 4일 전의 날짜까지 확인이 가능하다는점!~!
future.tail()
```

```python
# forecast라는 변수에 위에서 만든 앞으로 30일동안의 데이터를 담은 변수 future을 넣어주고
forecast = m.predict(future)

# 예측한 데이터를 확인해보면
# 이 중 가장 오른쪽에 있는 yhat이라는 컬럼의 값이 프로펫 모델이 예측한 삼성전자의 종가의 값인것을 알수 잇죠 드래그 해서 오른쪽 맨끝이에요~!
forecast.tail()
```

```python
# 이제 그래프의 값을 확인해봅니다
# 이 그래프에는 검은색의 점과 파란색의 선이 존재하는데, 의미는 아래와 같습니다>~!
# 검은색 점 : 실제 삼성전자의 종가
# 파란색 선 : 삼성전자의 종가 트렌드 (흐름)
fig = m.plot(forecast)

# 그래프의 우측 끝 쪽에는 검은색 점이 없고 파란색 선만 존재하는 1달치의 그래프가 있는데, 이 부분의 파란색 선이 바로 프로펫이 예측한 삼성전자의 종가를 나타내는 선입니다.
```

```python
# 현재로부터 지난 날들의 삼성전자 종가 또한 예측을 해볼 수있는 것이죠~!
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].iloc[-40:-20]
```

```python
# 인터렉티브 시각화 하는 코드
# 확대해서 보고 싶은 부분을 좌클릭 드래그로 확대해서 보실수 있어요~!
fig = plot_plotly(m, forecast)
fig
```

```python
# 이건 안 써도 되는데... 그냥 해봣어요....
plot_components_plotly(m, forecast)
```
